```python
import os
import chromadb
```

```python
# 현재 파일 기준으로 chroma_db 폴더 경로 잡기
DB_DIR = os.path.join(os.getcwd(), "chroma_db") 
client = chromadb.PersistentClient(path=DB_DIR)
collection = client.get_collection(name="image_rag_cosine")

# 1. DB에 저장된 총 이미지 개수 확인!
count = collection.count()
print(f"우가! DB에 총 {count}개의 이미지가 저장되어 있다!")

# 2. 저장된 데이터 전체 ID(파일 경로)와 메타데이터 싹 다 가져와서 구경하기!
all_data = collection.get()

print("\n[저장된 파일 목록 훔쳐보기]")
if count > 0:
    for i in range(len(all_data['ids'])):
        file_id = all_data['ids'][i]
        meta = all_data['metadatas'][i]
        print(f"{i+1}번 파일 경로: {file_id}")
        print(f"   -> 메타데이터: {meta}")
        print(f"   -> Document: {all_data['documents'][i]}")
else:
    print("텅텅 비었다 우가! rag_ingest.py 먼저 돌려라!")

# 3. 혹시나 임베딩 벡터 숫자가 궁금하면 peek()을 쓴다! (5개만 뽑기)
peek_data = collection.peek(5)
print(peek_data['embeddings'])
print(peek_data['embeddings'][0].shape)
```
